### Draw Accel

```python
import matplotlib.pyplot as plt
import numpy as np
import csv

file_list = [
    "acc_huata.csv",
    "acc_jiayi.csv",
    "acc_shuwei.csv",
    "acc_yudai.csv"
]

for file_path in file_list:
    # file_path = "acc_huata.csv"  # Replace with actual path

    accel=[]
    with open(file_path, mode="r") as file:
        csv_reader = csv.reader(file)
        next(csv_reader)  # Skip the header if there is one
        for row in csv_reader:
            accel.append(float(row[1]))  # Y values

    # Create the plot
    plt.figure(figsize=(20, 10))
    plt.plot(accel, label="Acceleration", color="blue", marker="", linewidth=0.5)

    # Draw a horizontal red line at y=5.0
    plt.axhline(y=5.0, color="red", linestyle="--", label="Uncomfort")
    plt.axhline(y=-5.0, color="red", linestyle="--")

    # Labels and title
    plt.xlabel("Index")
    plt.ylabel("Acceleration")
    title = "Acceleration Plot (" + str(file_path) + ")"
    plt.title(title)
    plt.legend()

    # Show the plot
    plt.grid(True)
    plt.show()
```
